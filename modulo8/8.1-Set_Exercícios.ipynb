### Q1. 

Dada uma string qualquer, use seu conhecimento de sets para apresentar os caracteres únicos que aparecem na string, ordenados alfabeticamente e sem duplicatas. Preste atenção em duplicatas de maiúsculas e minúsculas.


````python
frase = "O rato roeu a roupa do Robson"

## Implemente aqui sua solução
# 1. Converte a frase para minúsculas
frase_lower = frase.lower()

# 2. Cria um set para obter caracteres únicos, excluindo o espaço
caracteres_unicos = set()
for char in frase_lower:
    if char != ' ':
        caracteres_unicos.add(char)

# 3. Converte o set para lista e ordena alfabeticamente
caracteres_ordenados = sorted(list(caracteres_unicos))

print(f"Frase original: \"{frase}\"")
print(f"Caracteres únicos e ordenados: {caracteres_ordenados}")
````

### Q2

Desenvolva uma função em Python chamada ```tem_elementos_comuns()``` que recebe duas listas como parâmetros e retorna True se houver algum elemento comum entre as duas listas, e False caso contrário. Utilize sets para resolver essa tarefa.

````python
# Implemente aqui a função tem_elementos_comuns()
def tem_elementos_comuns(lista1, lista2):
    """
    Verifica se há elementos comuns entre duas listas usando sets.
    Retorna True se houver interseção, False caso contrário.
    """
    # Converte as listas em sets para usar a operação de interseção
    set1 = set(lista1)
    set2 = set(lista2)
    
    # A interseção (&) retorna um novo set com os elementos comuns.
    # Se o set resultante não estiver vazio, há elementos comuns.
    return bool(set1 & set2)

# Teste:
lista1 = [1, 2, 3, 4]
lista2 = [3, 4, 5, 6, 7]
lista3 = [8, 9, 10]

print(f"Comum entre {lista1} e {lista2}: {tem_elementos_comuns(lista1, lista2)}")  # Saída esperada: True
print(f"Comum entre {lista1} e {lista3}: {tem_elementos_comuns(lista1, lista3)}")  # Saída esperada: False
````

````python
# Execute esta célula para testar a função da Q2.

lista1 = [1, 2, 3, 4]
lista2 = [3, 4, 5, 6, 7]
resultado = tem_elementos_comuns(lista1, lista2)
print(resultado)  # Saída esperada: True
````

### Q4. 
Considere duas listas que são duplicatas uma da outra, exceto por um elemento, exemplo:
```
A = [1, 4, 5, 7, 9]
B = [4, 5, 7, 9]
```

Encontre e apresente o elemento diferente, informando também qual das listas está desfalcada. Para o exemplo acima, a saída esperada é
```
O elemento 1 está faltando na segunda lista
```

````python
## Implemente aqui sua solução
A = [1, 4, 5, 7, 9]
B = [4, 5, 7, 9]

set_A = set(A)
set_B = set(B)

# Diferença: elementos que estão em A, mas não em B
diferenca_A = set_A - set_B

# Diferença: elementos que estão em B, mas não em A
diferenca_B = set_B - set_A

if diferenca_A:
    # O elemento está em A, mas B está desfalcada
    elemento_diferente = list(diferenca_A)[0]
    print(f"O elemento {elemento_diferente} está faltando na segunda lista")
elif diferenca_B:
    # O elemento está em B, mas A está desfalcada
    elemento_diferente = list(diferenca_B)[0]
    print(f"O elemento {elemento_diferente} está faltando na primeira lista")
else:
    print("As listas são idênticas")
````

### Q5.

Panagramas são frases que contém todas as letras do alfabeto pelo menos uma vez. Crie uma função ```checa_panagrama``` que recebe uma string e retorna ```True``` caso seja um panagrama ou ```False```caso contrário. Use seu conhecimento de sets para solucionar essa questão.

Ex: 
```
Entrada: "The quick brown fox jumps over the lazy dog" 
Saída: É um panagrama
```

```
Entrada: "Python é uma linguagem de programação" 
Saída: Não é um panagrama
```

````python
## Implemente aqui sua solução
import string

def checa_panagrama(frase):
    # 1. Define o conjunto de todas as letras do alfabeto (a-z)
    # Assumimos o alfabeto inglês de 26 letras, conforme o exemplo.
    alfabeto = set(string.ascii_lowercase)
    
    # 2. Processa a string:
    # a) Converte para minúsculas
    # b) Cria um set contendo apenas as letras
    letras_na_frase = set()
    for char in frase.lower():
        if 'a' <= char <= 'z': # Filtra apenas as letras
            letras_na_frase.add(char)
            
    # 3. Checagem: Verifica se o set do alfabeto é um subconjunto
    # (ou é igual) ao set de letras da frase.
    if alfabeto.issubset(letras_na_frase):
        return True
    else:
        return False

# Exemplos de uso:
frase1 = "The quick brown fox jumps over the lazy dog" 
frase2 = "Python é uma linguagem de programação"

if checa_panagrama(frase1):
    print(f'Entrada: "{frase1}" \nSaída: É um panagrama')
else:
    print(f'Entrada: "{frase1}" \nSaída: Não é um panagrama')

print("\n---")

if checa_panagrama(frase2):
    print(f'Entrada: "{frase2}" \nSaída: É um panagrama')
else:
    print(f'Entrada: "{frase2}" \nSaída: Não é um panagrama')
````
